```
weather_features = pd.read_csv('weather_features_final.csv')
flight_features = pd.read_csv('flight_features_final.csv')
```

```
weather_features = weather_features.drop(columns = weather_features.columns[0])
flight_features = flight_features.drop(columns = flight_features.columns[0])
#weather_features = weather_features.rename(columns = {"airport":"Origin"})
```

```
weather_features = weather_features.drop(columns = "weatherCode")
```

```
flight_features
```

```
weather_features = weather_features.drop(columns = ["visibility"])
```

```
flight_features['CRSDepTimedup'] = flight_features['CRSDepTime']
flight_features['CRSArrTimedup'] = flight_features['CRSArrTime']
flight_features = flight_features.round({'CRSDepTimedup':-2})
flight_features = flight_features.round({'CRSArrTimedup':-2})
```

```
final_features = pd.merge(flight_features,weather_features,left_on = ["CRSDepTimedup","FlightDate","Origin"],right_on = ["time","FlightDate","airport"] )
```

```
final_features = pd.merge(final_features,weather_features,left_on = ["CRSArrTimedup","FlightDate","Dest"],right_on = ["time","FlightDate","airport"] )
```
